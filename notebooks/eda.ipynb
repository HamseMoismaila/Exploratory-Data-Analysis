```python
top10 = (
    df.groupby("Speed_Mph")["Gforce_clean"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

sns.barplot(x="Speed_Mph", y="Gforce_clean", data=top10)
plt.title("Top 10 Speeds by Average G-force")
plt.show()
```

```python
top10 = df["Speed_Mph"].value_counts().head(10)

sns.barplot(x=top10.index, y=top10.values)
plt.title("Top 10 Speed_Mph Counts")
plt.xlabel("Speed_Mph")
plt.ylabel("Count")
plt.show()


df.head()
```
